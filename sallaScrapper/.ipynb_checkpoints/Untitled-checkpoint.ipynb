```python
def ClearSheet(sheet):
    for i in range(3,9920+3):
        for j in range(1,31):
            sheet.cell(row=i,column=j).value=''
```

```python
def addFirstRow(FirstRowEx,newData):
    nameIndice=3
    catIndice=4
    picIndice=5
    priceIndice=7
    markIndice=20
    indices=[nameIndice,catIndice,picIndice,priceIndice,markIndice]
    for i in range(len(indices)):
        FirstRowEx[i]=newData[i]
    return FirstRowEx
```

```python
def addOption(optionEx,newData,image):
    nameIndice=3
    picIndice=5
    priceIndice=7
    if image:
        indices=[nameIndice,picIndice,priceIndice]
    else:
        indices=[nameIndice,priceIndice]
    for i in range(len(indices)):
        optionEx[i]=newData[i]
        
    return optionEx
```
